#### Grid Line

```python
import matplotlib.pyplot as plt
import pandas as pd
data={
    "Cricket_bat":["sg","BDM","SS","Super"],
    "MRP":[2000,2500,2800,3000],
    "Weight":[110,115,120,130]
}

df=pd.DataFrame(data)

plt.plot(df["MRP"],df["Weight"])
plt.grid()
plt.show()
```

#### Lables to a plot

```python
import matplotlib.pyplot as plt
import pandas as pd
data={
    "Cricket_bat":["sg","BDM","SS","Super"],
    "MRP":[2000,2500,2800,3000],
    "Weight":[110,115,120,130]
}

df=pd.DataFrame(data)

plt.plot(df["MRP"],df["Weight"])


plt.xlabel("Bat price(BDT)")
plt.ylabel("Bat Weight(Grams)")
plt.show()
```

#### Bar graph

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

student=np.array(["Akram","Hossain","Tanveer","Afrin"])
mark=np.array([85,90,75,80])

plt.bar(student,mark)
plt.xlabel("Student")
plt.ylabel("Marks")

plt.show()
```

#### Pie chart

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

student=np.array(["Akram","Hossain","Tanveer","Afrin"])
mark=np.array([85,90,75,80])

plt.pie(mark,labels=student,autopct='%1.2f%%')


plt.show()
```
